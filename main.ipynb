```python
url = "https://www.jebosseengrandedistribution.fr/2020/04/14/voici-la-liste-des-rayons-quon-trouve-en-grande-distribution/?srsltid=AfmBOooak7kkzniP0a2veGYmU5VTJTseh3MR_lb8eC8HllV50yvuwmuJ"
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text , 'html.parser')
soup
```

```python
post_content = soup.find('div', {'class': 'post-content'})
```

```python
children = post_content.find_all()
```

```python

data = { }


for i , child in enumerate(children) : 
    if child.name == 'h3': 
        ul = children[i+1]
        title = child.text
        
        if ul.name == 'ul' : 
            
            data[title] = []
            
            for li in ul.find_all() :
                data[title].append(li.text)
            
            
            
data
```

```python

with open('data.json', 'w', encoding='utf-8') as file:
    json.dump(data, file, ensure_ascii=False, indent=4)
```
